## Nombre de la clase: Electricista

```python
class Electricista:
    # Constructor: Aquí se definen los atributos
    def __init__(self, nombre, telefono, edad):
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    # Función 1: Instalar
    def instalar(self):
        print(f"{self.nombre} está instalando un panel eléctrico de seguridad.")

    # Función 2: Cablear
    def cablear(self):
        print(f"{self.nombre} está realizando el cableado interno de la vivienda.")

    # Función 3: Iluminar
    def iluminar(self):
        print(f"{self.nombre} ha terminado el trabajo. ¡Ahora hay luz en la habitación!")

# --- Creación de Objetos (Instancias) ---

# Objeto 1
electricista1 = Electricista("Laura Gómez", "555-0123", 28)

# Objeto 2
electricista2 = Electricista("Beatriz Ruiz", "555-0987", 35)

# --- Demostración de uso ---

print(f"Datos de la Electricista 1: {electricista1.nombre}, Tel: {electricista1.telefono}")
electricista1.instalar()
electricista1.cablear()
electricista1.iluminar()

print("-" * 30)

print(f"Datos de la Electricista 2: {electricista2.nombre}, Edad: {electricista2.edad}")
electricista2.iluminar()

print("Programa realizado por Cepeda Darrell NL = 10")
```
